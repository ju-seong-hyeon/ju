```python
for i in glob.glob(folder):
    x = i.split('/')
    x.reverse()
    name = x[0][4:]
    name = name.split(".")
    name = name[0]+'.png'
    
    f = open(i, 'r')
    rdr = csv.reader(f)
    #print(name)
    
    for line in rdr:
        if(line[0] == "MD_019"):
            continue
        #print(line)
        #print(line[0])
        poi_name.append(line[0])
        x_list.append([int(line[2]), int(line[4])])
        y_list.append([int(line[3]), int(line[5])])
        #print(poi_name)
        
    length = len(poi_name)
    
    file_data = {}
    file_data[name] = {"fileref" : "", "size" : "", "filename":name, "base64_img_data": "","file_attributes": {}, "regions":{count : {}}}
    
    for i in range(0, length):
        print(poi_name[i])
        
        k=str(poi_name[i])
        
        if(k=="DT_001"):
            p_name = "F1111"
        elif(k=="DT_002"):
            p_name = "F1112"
        elif(k=="DT_127"):
            p_name = "F2511"
        elif(k=="DT_131"):
            p_name = "F1322"
        elif(k=="DT_143"):
            p_name = "F2542"
        elif(k=="DT_150"):
            p_name = "F3342"
        elif(k=="DT_177"):
            p_name = "F6110"
        elif(k=="SM_100"):
            p_name = "F2240"
        elif(k=="SM_126"):
            p_name = "F3130"
        elif(k=="SM_142"):
            p_name = "F3330"
        
        file_data[name]["regions"][count]={"shape_attributes" :{"name":"polygon", "all_points_x":x_list[i],"all_points_y":y_list[i]},"region_attributes": {"name":p_name}}
        
        count+=1
        count1+=1 
            
        
    with open('C:/Users/USER/Desktop/joo/project/working/use_json/new4/data'+str(count1)+".json", 'w') as make_file:
        json.dump(file_data, make_file, ensure_ascii=False, indent = "\t")
        
    count = 0    
    f.close()
    poi_name.clear()
    x_list.clear()
    y_list.clear()
    

    
#print(poi_name)
#print(x_list)
#print(y_list)

```
