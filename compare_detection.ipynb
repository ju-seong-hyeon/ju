```python
count_dict = {
    'F1111':0,
    'F1112':0,
    'F2511':0,
    'F2516':0,
    'F2542':0,
    'F3342':0,
    'F6110':0,
    'F2240':0,
    'F3130':0,
    'F3330':0 }

name_dict = {
    'DT_001':'F1111',
    'DT_002':'F1112',
    'DT_127':'F2511',
    'DT_131':'F2516',
    'DT_143':'F2542',
    'DT_150':'F3342',
    'DT_177':'F6110',
    'SM_100':'F2240',
    'SM_126':'F3130',
    'SM_142':'F3330',
    }


foi_dict = {}
foi_dict1 = {}

tmp_dict ={}
tmp_dict1 = {}

foi_list = []

name = ""
```

```python
class foi_dict(object):
    def __init__(self,name):
        self.name = name
    def __str__(self):
        return self.name
    def __repr__(self):
        return "'"+self.name+"'"
    
class foi_dict1(object):
    def __init__(self,name):
        self.name = name
    def __str__(self):
        return self.name
    def __repr__(self):
        return "'"+self.name+"'"
    
    
    
    
```

```python
for i in folder_list:
    with open(folder+i, 'r') as f:
        reader = csv.DictReader(f)
                
        for c in reader:
            for k, v in c.items():
                if(k=='name'):
                    if(v == 'F1322'):
                        name = 'F2516'
                    foi_list.append(name)
                            
                elif(k=="x1"):
                    foi_list.append(v)
                            
                elif(k=="y1"):
                    foi_list.append(v)
                    tmp_dict[foi_dict(foi_list[0])] = [foi_list[1], foi_list[2]]
                    foi_list.clear()
                    
    if os.path.isfile(folder1+i):                        
        with open(folder1+i, 'r') as b:
            reader = csv.DictReader(b)
                
            for c in reader:
                for k, v in c.items():
                    if(k=='name'):
                        name = name_dict.get(str(v))
                        foi_list.append(name)
                            
                    elif(k=="x1"):
                        foi_list.append(v)
                            
                    elif(k=="y1"):
                        foi_list.append(v)
                        tmp_dict1[foi_dict1(foi_list[0])] = [foi_list[1], foi_list[2]]
                        foi_list.clear()
                
        for key, val in tmp_dict.items():
            for key1, val1 in tmp_dict1.items():
                if(key == key1):
                    if((val[0] == (val1[0]+1 or val1[0] or val1[0]-1)) and
                      (val[1] == (val1[1]+1 or val1[1] or val1[1]-1))):
                        print('1')
                        a = count_dict.get(str(key))
                        count_dict[a]+=1    

```
